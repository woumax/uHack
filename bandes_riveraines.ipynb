```python
path =  "/Users/max/Downloads/cours-eau/cours_eau.shp"


water = gpd.read_file(path)

water
```

```python
import folium 

m = folium.Map(location=[water.geometry.centroid.y.mean(), water.geometry.centroid.x.mean()], zoom_start=8, tiles='cartodbpositron')

# Add a GeoJson layer to the map
tooltip = folium.features.GeoJsonTooltip(fields=['MUNICIPALITY_NAME'], aliases=['Municipality :'])
folium.GeoJson(water[['geometry', 'MUNICIPALITY_NAME']].to_json(), tooltip=tooltip, name='geojson').add_to(m)

map
```
